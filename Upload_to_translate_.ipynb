```python
##image preprocessing,
import cv2
import numpy as np

def preprocess_page(pil_img):
    img = np.array(pil_img)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)


    gray = cv2.GaussianBlur(gray, (5,5), 0)

    thresh = cv2.threshold(
        gray, 0, 255,
        cv2.THRESH_BINARY + cv2.THRESH_OTSU
    )[1]

    return thresh
```

```python
import pytesseract
def bilingual_ocr(img):
    text = pytesseract.image_to_string(
        img,
        lang="mar+eng",
        config="--psm 6 --oem 3"
    )
    return text.strip()
```

```python
from typing_extensions import final
final_text = ""

for i, page in enumerate(pages):
    processed = preprocess_page(page)
    page_text = bilingual_ocr(processed)

    final_text += f"\n--- Page {i+1} ---\n"
    final_text += page_text + "\n"
print(final_text)
```

```python
import re
from googletrans import Translator
```

```python

def create_protected_map(text):
    """
    Identify and protect URLs, email addresses, and other non-translatable content
    """
    protected_map = {}

    # Find URLs
    url_pattern = r'https?://[^\s]+|www\.[^\s]+'
    urls = re.findall(url_pattern, text)

    # Find email addresses
    email_pattern = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b'
    emails = re.findall(email_pattern, text)

    # Find dates and numbers (simplified pattern)
    date_pattern = r'\b\d{1,2}[-/]\d{1,2}[-/]\d{2,4}\b|\b\d{4}[-/]\d{1,2}[-/]\d{1,2}\b'
    dates = re.findall(date_pattern, text)

    # Replace protected content with placeholders
    protected_text = text
    all_protected = urls + emails + dates

    for idx, item in enumerate(all_protected):
        placeholder = f"__PROTECTED_{idx}__"
        protected_map[placeholder] = item
        protected_text = protected_text.replace(item, placeholder)

    return protected_text, protected_map
```

```python
def restore_english(text, protected_map):
    """
    Restore protected content after translation
    """
    restored_text = text
    for placeholder, original in protected_map.items():
        restored_text = restored_text.replace(placeholder, original)
    return restored_text

def contains_marathi(text):
    """
    Check if text contains Marathi characters (Devanagari script)
    Marathi Unicode range: U+0900 to U+097F
    """
    for char in text:
        if '\u0900' <= char <= '\u097F':
            return True
    return False
```

```python
def translate_marathi_to_english(text):
    """
    Translate Marathi text to English using Google Translate
    """
    try:
        translator = Translator()
        # For better results with mixed content, split into sentences
        sentences = text.split('. ')
        translated_sentences = []

        for sentence in sentences:
            if sentence.strip() and contains_marathi(sentence):
                translation = translator.translate(sentence, src='mr', dest='en')
                translated_sentences.append(translation.text)
            else:
                translated_sentences.append(sentence)

        # Join sentences back
        result = '. '.join(translated_sentences)

        # Clean up common translation artifacts
        result = result.replace(' .', '.').replace(' ,', ',')

        return result
    except Exception as e:
        print(f"Translation error: {e}")
        return text  # Return original text if translation fails
```

```python
def safe_document_translation(text):
    """
    Main function to safely translate document while preserving structure
    """
    # Step 1: Protect non-translatable content
    protected_text, protected_map = create_protected_map(text)

    # Step 2: Split into lines for structured translation
    lines = protected_text.split('\n')
    translated_lines = []

    for line in lines:
        line = line.strip()

        if not line:
            translated_lines.append("")
            continue

        # Check if line contains Marathi
        if contains_marathi(line):
            try:
                translated = translate_marathi_to_english(line)
                translated_lines.append(translated)
            except Exception as e:
                print(f"Line translation error: {e}")
                translated_lines.append(line)
        else:
            # Keep English or metadata as-is
            translated_lines.append(line)

    # Step 3: Join lines and restore protected content
    translated_text = "\n".join(translated_lines)
    final_translated = restore_english(translated_text, protected_map)

    return final_translated
```

```python
translated_text = safe_document_translation(final_text)

print("\n" + "="*50)
print("Translated Text:")
print("="*50)
print(translated_text)
```
